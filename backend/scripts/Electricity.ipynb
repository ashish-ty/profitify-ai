```python
total_expense = 10000
```

```python
## scc_expense -> this will be from expense data 
total_load = df_drivers['connected_load'].sum()
df_drivers['scc_expense'] = df_drivers['connected_load'] / total_load * total_expense
```

```python
df_drivers
```

```python
## merging with SCC 
df_alloc = df_services.merge(df_drivers[['SCC_code','scc_expense']], on='SCC_code')
```

```python
## merge with services and service cost based on TAT 

## NOTE - this TAT can vary from patient to patient so we need to keep this tat on patient level then
tat_sum = df_alloc.groupby('SCC_code')['TAT'].transform('sum')
df_alloc['service_cost'] = df_alloc['scc_expense'] * df_alloc['TAT'] / tat_sum
```

```python
import pandas as pd
import numpy as np
import random
from datetime import datetime, timedelta

# Helper functions
def random_date(start, end):
    return start + timedelta(days=random.randint(0, (end - start).days))

# Settings
n = 100  # number of rows

# Predefined lists
cities = ['Mumbai', 'Delhi', 'Bengaluru', 'Chennai', 'Kolkata']
genders = ['M', 'F']
dept_codes = ['CARD', 'NEUR', 'ORTH', 'MED', 'SURG']
dept_names = {
    'CARD': 'Cardiology',
    'NEUR': 'Neurology',
    'ORTH': 'Orthopedics',
    'MED': 'Medicine',
    'SURG': 'Surgery'
}
bill_types = ['Regular', 'Emergency', 'Insurance']
patient_categories = ['General', 'Private', 'VIP']
payor_types = ['Self', 'Insurance']
insurance_names = ['Aetna', 'BlueCross', 'UnitedHealth', 'None']
payment_modes = ['Cash', 'Card', 'Online']
billing_categories = ['Inpatient', 'Outpatient']
service_status = ['Completed', 'Pending', 'Cancelled']
yes_no = [True, False]
sub_cost_centers = ['ASC0101', 'ASC0102', 'ASC0103', 'ASC0201', 'ASC0202']

# 1. Bill Register
bill_data = []
for i in range(n):
    admit = random_date(datetime(2025,1,1), datetime(2025,6,30))
    discharge = admit + timedelta(days=random.randint(1,10))
    uhid = f"UHID{10000+i}"
    ipd = f"IPD{20000+i}"
    bill = f"BILL{30000+i}"
    dept = random.choice(dept_codes)
    gross = round(random.uniform(5000, 50000), 2)
    discount = round(gross * random.uniform(0, 0.2), 2)
    net = gross - discount
    icu = random.randint(0, 5)
    nonicu = max(0, (discharge - admit).days - icu)
    copays = [round(net * random.uniform(0, 0.1), 2) for _ in range(4)]
    bill_data.append({
        'UHID': uhid,
        'IPD_No': ipd,
        'Bill_No': bill,
        'City_State': random.choice(cities),
        'Gender_Age': f"{random.choice(genders)}/{random.randint(1, 100)}",
        'Admission_Date': admit,
        'Discharge_Date': discharge,
        'Doctor_ID': f"DR{random.randint(100,200)}",
        'Dept_Code': dept,
        'Dept_Name': dept_names[dept],
        'Doctor_Name': f"Dr. {random.choice(['Smith','Patel','Lee','Khan','Singh'])}",
        'Bill_Type': random.choice(bill_types),
        'Bill_Month': admit.strftime('%Y-%m'),
        'Patient_Category': random.choice(patient_categories),
        'Gross_Bill': gross,
        'Discount_Amount': discount,
        'Net_Bill': net,
        'ICU_Days': icu,
        'NonICU_Days': nonicu,
        'Payor_Type': random.choice(payor_types),
        'Insurance_Name': random.choice(insurance_names),
        'Payment_Mode': random.choice(payment_modes),
        'Billing_Category': random.choice(billing_categories),
        'CoPay1': yes_no[0],
        'CoPay1_Amount': copays[0],
        'CoPay2': yes_no[1],
        'CoPay2_Amount': copays[1],
        'CoPay3': yes_no[0],
        'CoPay3_Amount': copays[2],
        'CoPay4': yes_no[1],
        'CoPay4_Amount': copays[3],
    })
df_bill = pd.DataFrame(bill_data)

# 2. Patient Service Register
psr_data = []
for i in range(n):
    admit = random_date(datetime(2025,1,1), datetime(2025,6,30))
    service_date = admit + timedelta(days=random.randint(0, (datetime(2025,6,30)-admit).days))
    dept = random.choice(dept_codes)
    service_code = f"SVC{random.randint(100,200)}"
    service_dept = random.choice(dept_codes)
    subcc = random.choice(sub_cost_centers)
    gross_amt = round(random.uniform(100, 5000),2)
    discount = round(gross_amt * random.uniform(0, 0.15),2)
    net_amt = gross_amt - discount
    tat = random.uniform(0.5, 5)  # hours or days
    psr_data.append({
        'SerialNo': i+1,
        'IPDNo': f"IPD{20000+i}",
        'AdmitDocID': f"DR{random.randint(100,200)}",
        'AdmitDeptCode': dept,
        'AdmitDeptName': dept_names[dept],
        'AdmitDocName': f"Dr. {random.choice(['Patel','Mehta','Shah','Roy','Verma'])}",
        'PerformDocID': f"DR{random.randint(100,200)}",
        'PerformDocName': f"Dr. {random.choice(['Sharma','Gupta','Rao','Das','Iyer'])}",
        'PerformDeptCode': service_dept,
        'PerformDeptName': dept_names[service_dept],
        'ReferringDocID': f"DR{random.randint(100,200)}",
        'ReferringDocName': f"Dr. {random.choice(['Desai','Joshi','Nair','Chowdhury','Mishra'])}",
        'ReferringDeptCode': dept,
        'ReferringDeptName': dept_names[dept],
        'FinalBillDate': discharge,
        'BillNo': f"BILL{30000+i}",
        'ServiceCode': service_code,
        'ServiceName': f"Service {random.randint(1,50)}",
        'ServiceDeptID': service_dept,
        'ServiceDeptName': dept_names[service_dept],
        'ServiceSubDeptID': subcc,
        'ServiceSubDeptName': f"SubCC {subcc}",
        'ServiceStatus': random.choice(service_status),
        'IsPackaged': random.choice(yes_no),
        'IsOutsourced': random.choice(yes_no),
        'Quantity': random.randint(1,10),
        'GrossAmount': gross_amt,
        'Discount': discount,
        'NetAmount': net_amt,
        'EmergencyChargeApplied': random.choice(yes_no),
        'PerformDocShare': round(net_amt * 0.5,2),
        'PharmaMaterialCost': round(net_amt * 0.2,2),
        'OutsourceShare': round(net_amt * 0.1,2),
        'SubCostCenterCode': subcc,
        'ServiceLocation': random.choice(cities),
        'ServiceDate': service_date,
        'TAT': tat
    })
df_psr = pd.DataFrame(psr_data)

# 3. Trial Balance
tb_data = []
for i in range(n):
    ledger = f"LED{300+i}"
    code = f"C{random.randint(100,200)}"
    date = random_date(datetime(2025,1,1), datetime(2025,6,30))
    debit = round(random.uniform(1000,10000),2)
    credit = round(random.uniform(1000,10000),2)
    net = debit - credit
    tb_data.append({
        'Code': code,
        'Ledger': ledger,
        'Alias': ledger.lower(),
        'Date': date,
        'Month': date.strftime('%Y-%m'),
        'Post ID': f"PST{400+i}",
        'Desc': random.choice(['Electricity','Laundry','Water','Medicines']),
        'Open Bal': round(random.uniform(0,5000),2),
        'Debit': debit,
        'Credit': credit,
        'Close Bal': round(random.uniform(0,5000),2),
        'Net Bal': net,
        'Notes': ''
    })
df_tb = pd.DataFrame(tb_data)

# 4. Power Consumption
pc_data = []
for i, subcc in enumerate(sub_cost_centers, start=1):
    conn = random.uniform(5,20)
    run = random.uniform(1,conn)
    stdby = conn - run
    days = 30
    hrs = 24
    total = conn * days * hrs
    pc_data.append({
        'No': i,
        'SubCCCode': subcc,
        'SubCC': f"SubCC {subcc}",
        'ConnLoad': round(conn,2),
        'RunLoad': round(run,2),
        'StdbyLoad': round(stdby,2),
        'Days': days,
        'Hrs': hrs,
        'TotalLoadKg': round(total,2),
        'Notes': ''
    })
df_pc = pd.DataFrame(pc_data)

# # Display to user
# tools.display_dataframe_to_user("Bill Register (Sample)", df_bill)
# tools.display_dataframe_to_user("Patient Service Register (Sample)", df_psr)
# tools.display_dataframe_to_user("Trial Balance (Sample)", df_tb)
# tools.display_dataframe_to_user("Power Consumption (Sample)", df_pc)
```

```python
import pandas as pd
import numpy as np
import random
from datetime import datetime, timedelta

# Total electricity cost -> Confirm with Bhaiya 
total_elec_cost = df_tb[df_tb.Desc == 'Electricity'][df_tb.Month == '2025-04']['Debit'].sum()

# Allocate to each SCC by connected load
df_pc['scc_expense'] = df_pc['ConnLoad'] / df_pc['ConnLoad'].sum() * total_elec_cost

# Merge SCC expense into PSR
df_alloc = df_psr.merge(df_pc[['SubCCCode','scc_expense']], 
                        left_on='SubCostCenterCode', right_on='SubCCCode')

# Within each SCC, allocate by TAT
tat_sum = df_alloc.groupby('SubCostCenterCode')['TAT'].transform('sum')
df_alloc['service_cost'] = df_alloc['scc_expense'] * df_alloc['TAT'] / tat_sum

# Final columns
df_final = df_alloc[['SerialNo','IPDNo','ServiceCode','ServiceName',
                     'SubCostCenterCode','TAT','scc_expense','service_cost']]

```
